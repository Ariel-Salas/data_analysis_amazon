1. df.isnull()

    df.isnull() genera un DataFrame de valores booleanos (True/False) donde:
        True indica que en esa celda del DataFrame hay un valor nulo o faltante (NaN).
        False indica que la celda tiene un valor presente (no nulo).

    Este método devuelve un DataFrame del mismo tamaño que el original, pero con valores True y False según la presencia de valores nulos.

2. .sum()

    Al aplicar .sum() después de df.isnull(), estamos sumando los valores True en cada columna. En Python, True se considera como 1 y False como 0, por lo que cuando sumamos los valores de True, obtenemos el número total de valores nulos en cada columna.
        Por ejemplo, si una columna tiene 2 valores True (nulos) y 10 valores False (no nulos), el resultado para esa columna será 2. Que es justamente lo que pasa en rating_count que tiene el valor de 2

```python

# Importar las bibliotecas necesarias
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

# Cargar el dataset
file_path = "C:/Users/ariel/OneDrive/Desktop/ciencia de datos/dataset/amazon.csv"
df = pd.read_csv(file_path)

# Limpieza de datos: transformar columnas a formatos numéricos
df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace(',', '').astype(float)
df['discounted_price'] = df['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')  # Convertir ratings a números, ignorando errores

# Filtrar valores nulos y anómalos
df_cleaned = df.dropna(subset=['actual_price', 'discounted_price', 'rating'])

# Crear la gráfica interactiva en 3D
fig = px.scatter_3d(
    df_cleaned,
    x='actual_price',       # Eje X: Precio original
    y='discounted_price',   # Eje Y: Precio con descuento
    z='rating',             # Eje Z: Calificación
    color='rating',         # Colorear por calificación
    hover_name='product_name',  # Mostrar nombre del producto al pasar el mouse
    title='Relación entre Precios y Calificaciones de Productos'
)

# Ajustar el tamaño del gráfico y los puntos
fig.update_layout(
    width=1200,  # Ancho del gráfico
    height=800,  # Altura del gráfico
    scene=dict(
        xaxis=dict(title='Precio Original', color='red'),  # Eje X en rojo
        yaxis=dict(title='Precio Descuento', color='green'),  # Eje Y en verde
        zaxis=dict(title='Calificación', color='blue')  # Eje Z en azul
    )
)

# Reducir el tamaño de los puntos
fig.update_traces(marker=dict(size=3))  # Tamaño de los puntos reducido

# Agregar flechas a los ejes
fig.add_trace(
    go.Scatter3d(
        x=[df_cleaned['actual_price'].max()], 
        y=[0], 
        z=[0], 
        mode='markers+text',
        marker=dict(size=5, color='red'),
        text=['Eje X →'],
        textposition="middle right",
        showlegend=False
    )
)
fig.add_trace(
    go.Scatter3d(
        x=[0], 
        y=[df_cleaned['discounted_price'].max()], 
        z=[0], 
        mode='markers+text',
        marker=dict(size=5, color='green'),
        text=['Eje Y →'],
        textposition="middle right",
        showlegend=False
    )
)
fig.add_trace(
    go.Scatter3d(
        x=[0], 
        y=[0], 
        z=[df_cleaned['rating'].max()], 
        mode='markers+text',
        marker=dict(size=5, color='blue'),
        text=['Eje Z →'],
        textposition="middle right",
        showlegend=False
    )
)

# Mostrar la gráfica
fig.show()

```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Cargar el dataset
file_path = "C:/Users/ariel/OneDrive/Desktop/ciencia de datos/dataset/amazon.csv"
df = pd.read_csv(file_path)

# Limpieza de datos: transformar columnas a formatos numéricos
df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace(',', '').astype(float)
df['discounted_price'] = df['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')  # Convertir ratings a números, ignorando errores

# Filtrar valores nulos y anómalos
df_cleaned = df.dropna(subset=['actual_price', 'discounted_price', 'rating'])

# Crear la gráfica interactiva en 3D
fig = px.scatter_3d(
    df_cleaned,
    x='actual_price',       # Eje X: Precio original
    y='discounted_price',   # Eje Y: Precio con descuento
    z='rating',             # Eje Z: Calificación
    color='rating',         # Colorear por calificación
    hover_name='product_name',  # Mostrar nombre del producto al pasar el mouse
    title='Relación entre Precios y Calificaciones de Productos'
)

# Ajustar el tamaño del gráfico y los puntos
fig.update_layout(
    width=1200,
    height=800,
    scene=dict(
        xaxis=dict(title='Precio Original'),
        yaxis=dict(title='Precio Descuento'),
        zaxis=dict(title='Calificación')
    )
)

# Agregar botones para cambiar vistas
fig.update_layout(
    updatemenus=[
        dict(
            type="buttons",
            direction="left",
            x=0.1,
            y=1.2,
            buttons=[
                dict(
                    label="Vista Frontal",
                    method="relayout",
                    args=["scene.camera", {
                        "eye": {"x": 0, "y": 0, "z": 2.5}  # Frontal
                    }]
                ),
                dict(
                    label="Vista Lateral",
                    method="relayout",
                    args=["scene.camera", {
                        "eye": {"x": 2.5, "y": 0, "z": 0}  # Lateral
                    }]
                ),
                dict(
                    label="Vista Superior",
                    method="relayout",
                    args=["scene.camera", {
                        "eye": {"x": 0, "y": 2.5, "z": 0}  # Superior
                    }]
                ),
            ]
        )
    ]
)

# Mostrar la gráfica
fig.show()
```
